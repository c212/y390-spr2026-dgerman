```python
import qiskit
```

```python
from qiskit import *
```

```python
from qiskit_aer import AerSimulator
```

```python
circuit = QuantumCircuit(11, 10)
circuit.x(10)
circuit.barrier()
circuit.h(range(10))
circuit.h(10)
circuit.barrier()
circuit.cx(8, 10)
circuit.cx(9, 10)
circuit.cx(7, 10)
circuit.cx(0, 10)
circuit.cx(4, 10)
circuit.barrier()
circuit.h(range(10))
circuit.barrier()
circuit.measure(range(10), range(10))

circuit.draw(output='mpl')
```

```python
device = AerSimulator()
job = device.run(circuit, shots = 1)
result = job.result()
counts = result.get_counts(circuit)
```

```python
print(counts)
```
