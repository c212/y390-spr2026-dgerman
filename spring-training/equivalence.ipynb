```python
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(2)
```

```python
qc.h(0)
qc.h(1)
qc.cx(0, 1)
qc.h(0)
qc.h(1)
display(qc.draw('mpl'))
```

```python
from qiskit.quantum_info import Operator
Operator.from_circuit(qc).draw(output='latex')
```
