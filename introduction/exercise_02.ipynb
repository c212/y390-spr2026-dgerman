```python
from qiskit import QuantumCircuit
```

```python
import numpy as np
import random

def generate_amplitudes():
    alpha = np.sqrt(random.uniform(0, 1))
    beta = np.sqrt(1 - alpha**2)
    return alpha, beta
```

```python
(a, b) = generate_amplitudes()
print(a, b)
```

```python
qc = QuantumCircuit(1)
qc.initialize((a, b))
display(qc.draw('mpl', initial_state=True))
```

```python
from qiskit.quantum_info import Statevector
a = Statevector.from_instruction(qc)
a.draw('latex')
```
