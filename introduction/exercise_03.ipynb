```python
from qiskit import QuantumCircuit
```

```python
import numpy as np
import random
a = np.sqrt(random.uniform(0, 1))
b = np.sqrt(1 - a**2)

qc = QuantumCircuit(3)
qc.initialize([a, b], 0, 0)
qc.barrier()
qc.h(2)
qc.cx(2, 1)
qc.barrier()
qc.cx(0, 1)
qc.h(0)
qc.barrier()
display(qc.draw('mpl'))
```

```python
from qiskit.quantum_info import Statevector
sv = Statevector.from_instruction(qc)
sv.draw('latex')
```

```python
result, state = sv.measure([0, 1])
print(f"Measured: {result}\nState after measurement:")
display(state.draw("latex"))
```
