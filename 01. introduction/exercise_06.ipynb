```python
from qiskit import QuantumCircuit, transpile
```

```python
import numpy as np
import random

def generate_amplitudes():
    alpha = (1-1j)/2
    beta = (1+1j)/2
    return alpha, beta
```

```python
generate_amplitudes()
```

```python
(a, b) = generate_amplitudes()
print(a, b)
```

```python
qc = QuantumCircuit(1, 1)
qc.initialize((a, b))
qc.barrier()
display(qc.draw('mpl', initial_state=True))
```

```python
from qiskit.quantum_info import Statevector
sv = Statevector.from_instruction(qc)
sv.draw('latex')
```

```python
qc.measure(range(1), range(1))

from qiskit_aer import AerSimulator
aersim = AerSimulator()
qc = transpile(qc, aersim)
```

```python
result = aersim.run(qc).result()
counts = result.get_counts(qc)
print(result.get_counts(qc))
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts, title='This is Exercise 06')
```
